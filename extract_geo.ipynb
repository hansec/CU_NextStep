```python
import ezdxf
import matplotlib.pyplot as plt
import numpy as np
```

```python
dxf = ezdxf.readfile('lim.dxf')
```

```python
#vv between 250 and 300
# lim between 300 and 350
```

```python
#320, div uppper
# 322, div lower
#325, 327 first wall
#[272:330]
modelspace = dxf.modelspace() 
polylines = modelspace.query('POLYLINE')
for entity in polylines:
    points = entity.points()
    contour = []
    for p in points:
        contour.append([p[0], p[1]])
    contour = np.array(contour)
    plt.plot(contour[:,0], contour[:,1], color = 'k', linewidth = 0.1)
ind = 271
for entity in polylines[ind:ind+1]:
    points = entity.points()
    contour = []
    for p in points:
        contour.append([p[0], p[1]])
    contour = np.array(contour)
    plt.fill(contour[:,0], contour[:,1], color = 'r')
    print(repr(contour))
plt.gca().set_aspect(aspect = 1)
```
