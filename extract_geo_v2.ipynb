```python
import ezdxf
import matplotlib.pyplot as plt
import numpy as np
```

```python
dxf_old = ezdxf.readfile('lim.dxf')
dxf_new = ezdxf.readfile('Section_05_07_24.dxf')
```

```python
#320, div uppper
# 322, div lower
#325, 327 first wall
#[272:330]
modelspace = dxf_new.modelspace() 
polylines = modelspace.query('POLYLINE')
for entity in polylines:
    points = entity.points()
    contour = []
    for p in points:
        contour.append([p[0], p[1]])
    contour = np.array(contour)
    plt.plot(contour[:,0], contour[:,1], color = 'k', linewidth = 0.1)

ind = 19
for entity in polylines[ind:ind+1]:
    points = entity.points()
    contour = []
    for p in points:
        contour.append([p[0], p[1]])
    contour = np.array(contour)
    plt.fill(contour[:,0], contour[:,1], color = 'r')

modelspace = dxf_old.modelspace() 
polylines = modelspace.query('POLYLINE')
for entity in polylines:
    points = entity.points()
    contour = []
    for p in points:
        contour.append([p[0], p[1]])
    contour = np.array(contour)
    plt.plot(contour[:,0], contour[:,1], color = 'r', linewidth = 0.1)


    
plt.gca().set_aspect(aspect = 1)
```
